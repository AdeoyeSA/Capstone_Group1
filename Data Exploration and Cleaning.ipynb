```python
import pandas as pd
import os

# 1. Define the data folder name
DATA_FOLDER = 'RealEstateData2000-2025'

# 2. Create the full paths to your specific files
# Using os.path.join makes this work on both Windows and Mac/Linux
active_path = os.path.join(DATA_FOLDER, 'RealEstateDataJanuary2026-Data3960-Active.csv')
sold_path = os.path.join(DATA_FOLDER, 'RealEstateDataJanuary2026-Data3960-Sold.csv')

# 3. Load the data
try:
    # Adding encoding='latin1' usually fixes the "0xe2" error
    active_df = pd.read_csv(active_path, encoding='latin1')
    sold_df = pd.read_csv(sold_path, encoding='latin1')
    
    print("â Success! Both datasets loaded.")
    print(f"Active Assets: {active_df.shape[0]} rows")
    print(f"Sold Assets: {sold_df.shape[0]} rows")
    
except FileNotFoundError:
    print("â Error: Could not find the files.")
except UnicodeDecodeError:
    # If latin1 fails, cp1252 is the next most likely candidate for Windows files
    active_df = pd.read_csv(active_path, encoding='cp1252')
    sold_df = pd.read_csv(sold_path, encoding='cp1252')
    print("â Success! Loaded using Windows-1252 encoding.")
```

```python

# Data Dictionary (Pipeline)

def create_data_dictionary(df, name):
    print(f"Data Dictionary for {name} Listings:")
    data_dict = pd.DataFrame({
        "Column": df.columns,
        "Data Type": df.dtypes,
        "Missing Values": df.isnull().sum(),
        "Unique Values": df.nunique()
    })
    print(data_dict)
    print("\nExplanation: Each column's type, number of missing values, and unique values are shown. This helps understand the dataset structure and identify fields needing attention.\n")
    return data_dict

active_dict = create_data_dictionary(active_df, "Active")
sold_dict = create_data_dictionary(sold_df, "Sold")
```
